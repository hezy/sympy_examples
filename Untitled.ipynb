```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.arange(0,101,1)
nums
```

```python
new_nums = nums[(nums%2 != 0) * (nums%3 == 0)]
new_nums
```

```python
np.sum(new_nums)
```

```python
np.cumsum(new_nums)
```

```python
theta = np.linspace(0, 2*np.pi, 1000)
r = 1 + 3/4 * np.sin(3*theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x,y)
```

```python
dA = 1/2 * r**2 * (theta[1] - theta[0])
sum(dA)
```

```python
dL = np.sqrt(r**2 + np.gradient(r,theta)**2) * (theta[1] - theta[0])
sum(dL)
```

```python
plt.plot(theta,r)
plt.plot(theta,np.gradient(r,theta))
plt.plot(theta,np.cumsum(dA))
```
